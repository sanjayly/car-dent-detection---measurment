```python
from ultralytics import YOLO

# Load lightweight model (better for small dataset)
model = YOLO("yolov8n.pt")

model.train(
    data="/content/dataset/dataset/data.yaml",
    epochs=40,
    imgsz=416,
    batch=8,          # safe for T4 GPU
    device=0,         # GPU
    workers=2,        # small parallelism (Colab safe)
    patience=0,

    # Light + effective augmentation (not heavy)
    degrees=5,
    translate=0.05,
    scale=0.3,
    fliplr=0.5,
    mosaic=0.5,
    mixup=0.0,
    hsv_h=0.015,
    hsv_s=0.5,
    hsv_v=0.3
)
```
